````julia
## Integracion en un solo paso, para ser utilizadas dentro de solveODE, ODEunidim y ODEmultidim

function Euler(f,y0,t0,h)
    return y0 + h*f(y0,t0)
end
function Euler(f,y0,t0,h,p)
    return y0 + h*f(y0,t0,p)
end

function RK2(f, y0, t0, h)
    k1 = h * f(y0, t0)
    k2 = h * f(y0 + k1, t0 + h)
    y1 = y0 + 0.5 * (k1 + k2)
    return y1
end
function RK2(f, y0, t0, h, p)
    k1 = h * f(y0, t0, p)
    k2 = h * f(y0 + k1, t0 + h, p)
    y1 = y0 + 0.5 * (k1 + k2)
    return y1
end

function RK4(f, y0, t0, h)
    k1 = h * f(y0, t0)
    k2 = h * f(y0 + 0.5 * k1, t0 + 0.5 * h)
    k3 = h * f(y0 + 0.5 * k2, t0 + 0.5 * h)
    k4 = h * f(y0 + k3, t0 + h)
    y1 = y0 + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
    return y1
end
function RK4(f, y0, t0, h, p)
    k1 = h * f(y0, t0, p)
    k2 = h * f(y0 + 0.5 * k1, t0 + 0.5 * h, p)
    k3 = h * f(y0 + 0.5 * k2, t0 + 0.5 * h,  p)
    k4 = h * f(y0 + k3, t0 + h, p)
    y1 = y0 + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
    return y1
end
````

````julia
## ODEunidim

function ODEunidim(f, method, y0, t0, h, a, b)
    
    num_steps = Int(round((b-a)/h))  # Número de pasos de integración
    t_values = [t0 + i * h for i in 0:num_steps]  # Vector de tiempo t
    w_values = zeros(num_steps+1)  # Vector de valores aproximados w
    w_values[1] = y0

    # Iteración para realizar los pasos de integración
    for i in 1:num_steps
        t_i = t_values[i]
        w_i = w_values[i]
        w_iplus1 = method(f, w_i, t_i, h)
        w_values[i+1] = w_iplus1
    end
    
    return t_values, w_values
end
function ODEunidim(f, method, y0, t0, h, a, b, p)
    
    num_steps = Int(round((b-a)/h))  # Número de pasos de integración
    t_values = [t0 + i * h for i in 0:num_steps]  # Vector de tiempo t
    w_values = zeros(num_steps+1)  # Vector de valores aproximados w
    w_values[1] = y0

    # Iteración para realizar los pasos de integración
    for i in 1:num_steps
        t_i = t_values[i]
        w_i = w_values[i]
        w_iplus1 = method(f, w_i, t_i, h, p)
        w_values[i+1] = w_iplus1
    end
    
    return t_values, w_values
end
````

````julia
"""ODEmultidim
pasar f como array de funciones
ejemplo:  f(y,t,p) = [y[2], -p[1]/p[2] * sin(y[1])]
"""
function ODEmultidim(f, method, y0, t0, h, a, b)
    N = Int(round((b-a)/h))
    n_eqs = length(y0)
    t = zeros(N)
    w = zeros(n_eqs,N)
    
    t[1]   = a
    w[:,1] = y0
    
    for i in 2:N
        t[i]   = t[i-1]+h
        w[:,i] = method(f,w[:,i-1],t[i-1],h)
    end
    return t[:], w[:,:]
end
function ODEmultidim(f, method, y0, t0, h, a, b, p)
    N = Int(round((b-a)/h))
    n_eqs = length(y0)
    t = zeros(N)
    w = zeros(n_eqs,N)
    
    t[1]   = a
    w[:,1] = y0
    
    for i in 2:N
        t[i]   = t[i-1]+h
        w[:,i] = method(f,w[:,i-1],t[i-1],h,p)
    end
    return t[:], w[:,:]
end
````

````julia
function solveODE(f, method, y0, t0, h, a, b)
    
    num_steps = Int(round((b-a)/h))  # Número de pasos de integración
    t_values = [t0 + i * h for i in 0:num_steps]  # Vector de tiempo t
    w_values = zeros(num_steps+1)  # Vector de valores aproximados w
    w_values[1] = y0
    
    # Selección del método de integración
    if method == "euler"
        method = euler_step
    elseif method == "rk2"
        method = rk2_step
    elseif method == "rk4"
        method = rk4_step
    end

    # Iteración para realizar los pasos de integración
    for i in 1:num_steps
        t_i = t_values[i]
        w_i = w_values[i]
        w_iplus1 = method(f, w_i, t_i, h)
        w_values[i+1] = w_iplus1
    end
    
    return t_values, w_values
end
function solveODE(f, method, y0, t0, h, a, b, p)
    
    num_steps = Int(round((b-a)/h))  # Número de pasos de integración
    t_values = [t0 + i * h for i in 0:num_steps]  # Vector de tiempo t
    w_values = zeros(num_steps+1)  # Vector de valores aproximados w
    w_values[1] = y0
    
    # Selección del método de integración
    if method == "euler"
        method = euler_step
    elseif method == "rk2"
        method = rk2_step
    elseif method == "rk4"
        method = rk4_step
    end

    # Iteración para realizar los pasos de integración
    for i in 1:num_steps
        t_i = t_values[i]
        w_i = w_values[i]
        w_iplus1 = method(f, w_i, t_i, h, p)
        w_values[i+1] = w_iplus1
    end
    
    return t_values, w_values
end
````

````julia
function rk4_multi(f, y0, t0, h, a, b; p = nothing)
    
    n_eqs = length(y0)
    
    n_steps = Int(round((b-a)/h))  # Número de pasos de integración
    
    t = [t0 + i*h for i in 0:n_steps]  # Vector de tiempo t
    w = zeros(n_eqs,n_steps+1)  # Vector de valores aproximados w
    w[:,1] = y0

    k1 = zeros(n_eqs)
    k2 = zeros(n_eqs)
    k3 = zeros(n_eqs)
    k4 = zeros(n_eqs)
    
    # Iteración para realizar los pasos de integración
    for i in 1:n_steps
        for j in 1:n_eqs
            k1[j] = h*f(w[:,i], t[i]; p)[j]
        end
        
        for j in 1:n_eqs
            k2[j] = h*f(w[:,i] .+ 0.5*k1, t[i] + 0.5*h; p)[j]
        end
        
        for j in 1:n_eqs
            k3[j] = h*f(w[:,i] .+ 0.5*k2, t[i] + 0.5*h; p)[j]
        end
        
        for j in 1:n_eqs
            k4[j] = h*f(w[:,i] .+ k3, t[i] + h; p)[j]
        end
        
        for j in 1:n_eqs
            w[j,i+1] = w[j,i] + (k1[j]+2*k2[j]+2*k3[j]+k4[j])/6
        end
    end
 
    return t, w
end
````

````julia
g, l = 9.8, 1.
t0 = 0.
y0 = [3.5,0.]
h = 0.01
a,b = 0., 10.
function f(y,t;p=[1.,1.])
    y1, y2 = y
    f1(y,t) = y2
    f2(y,t) = -p[1]/p[2] * sin(y1)
    return f1(y,t),f2(y,t)
end

sol4 = rk4_multi(f, y0, t0, h, a, b, p=[9.8,1.])

theta = sol4[2][1,:]
v = sol4[2][2,:]
t = sol4[1]

p1 = plot(t,theta,label=L"θ(t)")
p2 = plot(t,v,label=L"v(t)")
plot(p1,p2,layout=(2,1),xlabel=L"t")
````

Sea $S=$ población de individuos susceptibles, $I=$ población de individuos infectados y $R=$ población de individuos recuperados (que tienen inmunidad y no pueden volver a contagiarse).
Entonces las ecuaciones diferenciales asociadas a la evolución temporal de dichas poblaciones son:

\begin{align}
\frac{d S}{d t} &= -\beta S I /N\\
\frac{d I}{d t} &= \beta S I /N - \gamma I\\
\frac{d R}{d t} &= \gamma I
\end{align}

donde $N=S+I+R$ es la población total constante.

1. Considere las condiciones iniciales: $N=1000$, $I_0=10$, $R_0=0$ y los parámetros $\gamma = 0.1$ y  $\beta = 0.5$.
 
2. Utilice el método RK4 implementado en el **problema 4** para resolver este problema entre $t=0$ y $t=200$. 
 
3. Grafique la evolución temporal de las tres poblaciones, $S(t)$, $I(t)$ y $R(t)$ vs $t$ en el intervalo de $t$ calculado.

````julia
function prob5(y,t;p=nothing)
    S, I, R = y
    N = S + I + R
    β,γ = p[1], p[2]
    
    f1(y,t,p) = -β*S*I/N
    f2(y,t,p) = β*S*I/N - γ*I
    f3(y,t,p) = γ*I
    
    return f1(y,t,p), f2(y,t,p), f3(y,t,p)
end
````

### Rta. C.2.2

**Teoría**

Primero convertimos el problema de valor inicial de segundo orden a uno de primer orden. Sea $y_1=x$ y $y_2=x'$. Luego, $y_1'=x'=y_2$, $y_2'=x''=2x^3=2y_1^3$, $y_1(t_0)=x(t_0)=x_0$ y $y_2(t_0)=x'(t_0)=a$. Escribimos la solucion $x_a$ como $y_{a,1}$ y $F(a)=y_{a,1}(t_1)-x_1$.

````julia
function biseccion(f,a,b;epsx=1e-5,epsf=1e-5,n=100)
    fa = f(a)
    fb = f(b)
    
    if fa*fb > 0
        error("no hay raiz en [a,b]") 
    end
    
    for i in 1:n
        x = 0.5*(a+b)
        fx = f(x)
        if fa*fb >= 0
            a  = x
            fa = fx
        else
            b  = x
        end
        R = abs(b-a)/abs(b+a)
        F = abs(fx)
        if R < epsx && F < epsf
            return x, fx, R, F, i
        end
    end
    error("no converge")
end
````

````julia
function ODEmultidim(method, f, y0, a, b, N)
    h = (b-a)/(N-1)
    t = zeros(N)
    w = zeros((length(y0),N))
    
    t[1] = a
    w[:,1] = y0
    
    for i in 2:N
        t[i] = t[i-1]+h
        w[:,i] = method(f,w[:,i-1],t[i-1],h)
    end
    return t[:], w[:,:]
end
function ODEmultidim(method, f, y0, a, b, N, p)
    h = (b-a)/(N-1)
    t = zeros(N)
    w = zeros((length(y0),N))
    
    t[1] = a
    w[:,1] = y0
    
    for i in 2:N
        t[i] = t[i-1]+h
        w[:,i] = method(f,w[:,i-1],t[i-1],h,p)
    end
    return t[:], w[:,:]
end
````
